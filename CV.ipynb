```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
from pylab import rcParams
from pylab import figure
from pandas import read_excel 
matplotlib.rcParams['font.family'] = 'franklin gothic medium' 
plt.rcParams["font.style"] = 'italic'
matplotlib.rcParams['figure.figsize'] = 5,5
```

```python
data = r"AO-I-063-30-mM-CTAB-Soln-CV.xlsx"
CV = pd.read_excel(data, sheet_name=0, skiprows=[0],
                        usecols='A,B', 
                        names=['Voltage','Current'])
```

```python
x1 = CV['Voltage']
y1 = CV['Current']
```

```python
plt.plot(x1, y1,'b')
plt.ylabel('Current (A)',fontsize=16,weight='bold')
plt.xlabel('Voltage (V vs. SCE)',fontsize=16,weight='bold')

plt.tight_layout()
plt.subplots_adjust(bottom=0.15)
plt.subplots_adjust(left=0.10)
plt.savefig('AO-I-063-30-mM-CTAB-Soln-CV.jpg',dpi=300,bbox_inches='tight')
```
