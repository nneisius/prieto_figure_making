```python
file1 = '/Users/ngimble16/Documents/Colorado State University/Research/XPS/2021 FEC controls/02042021 Na soaked in FEC then uncycled in 111 swagelok with Cu2Sb/Na soaked in FEC then uncycled in 111 swagelok with Cu2Sb.xlsx'
file2 = '/Users/ngimble16/Documents/Colorado State University/Research/XPS/2021 FEC controls/03222021 Na soaked in FEC then uncycled in 111 2/Na in FEC then uncycled 111 2.xlsx'
file3 = '/Users/ngimble16/Documents/Colorado State University/Research/XPS/2021 FEC controls/04192021 Na FEC pre soak 3/Na in FEC 3.xlsx'

#extra wide range of binding energies data is within
BEmin = 270
BEmax = 305
alldata = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1))) 

locations=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file1, 1, 'B,C', 3], 
    [file1, 2, 'B,C', 3], 
    [file1, 3, 'B,C', 3], 
]

for num, param in enumerate(locations):
    data = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data = data.dropna()
    data['CPS'+str(num)] -= (data['CPS'+str(num)].min())
    plt.plot(data['BE'],data['CPS'+str(num)], color='grey', linewidth=1, label='Cu2Sb')
    
    data2 = data.set_index(round(data['BE'],1))
    data2 = data2.drop('BE',axis=1)
    alldata = pd.concat([alldata,data2], axis=1, join='inner')

alldata['avg'] = alldata.mean(axis=1)
plt.plot(alldata.index.values,alldata['avg'], color='black', linewidth=2)

alldata1 = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1))) 

locations1=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file2, 1, 'B,C', 3], 
    [file2, 2, 'B,C', 3], 
    [file2, 3, 'C,D', 3], 
]

for num, param in enumerate(locations1):
    data1 = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data1 = data1.dropna()
    data1['CPS'+str(num)] -= (data1['CPS'+str(num)].min())
    plt.plot(data1['BE'],data1['CPS'+str(num)], color='cyan', linewidth=1, label='Cu2Sb2')
    
    data3 = data1.set_index(round(data1['BE'],1))
    data3 = data3.drop('BE',axis=1)
    alldata1 = pd.concat([alldata1,data3], axis=1, join='inner')

alldata1['avg'] = alldata1.mean(axis=1)
plt.plot(alldata1.index.values,alldata1['avg'], color='blue', linewidth=2)

alldata2 = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1))) 

locations2=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file3, 1, 'B,C', 3], 
    [file3, 2, 'B,C', 3], 
    [file3, 3, 'B,C', 3], 
]

for num, param in enumerate(locations2):
    data2 = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data2 = data2.dropna()
    data2['CPS'+str(num)] -= (data2['CPS'+str(num)].min())
    plt.plot(data2['BE'],data2['CPS'+str(num)], color='orange', linewidth=1, label='Cu2Sb3')
    
    data4 = data2.set_index(round(data2['BE'],1))
    data4 = data4.drop('BE',axis=1)
    alldata2 = pd.concat([alldata2,data4], axis=1, join='inner')

alldata2['avg'] = alldata2.mean(axis=1)
plt.plot(alldata2.index.values,alldata2['avg'], color='red', linewidth=2)

plt.xlabel('Binding Energy (eV)',fontsize=22)
plt.ylabel('Subtracted Counts (a.u.)',fontsize=22)
plt.xlim(293,280)
plt.ylim(0,1600)
plt.xticks(fontsize=18, rotation=0)
plt.yticks(fontsize=18, rotation=0)
plt.legend(fontsize=18)
plt.show()
```
